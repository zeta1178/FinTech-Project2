```python
#use fincial modeling prpep api to get recession percentages
try:
    # For Python 3.0 and later
    from urllib.request import urlopen
except ImportError:
    # Fall back to Python 2's urllib2
    from urllib2 import urlopen
```

```python
import pandas as pd
import certifi
import json
from pathlib import Path
%matplotlib inline
```

```python
def get_jsonparsed_data(url):
    response = urlopen(url, cafile=certifi.where())
    data = response.read().decode("utf-8")
    return json.loads(data)
```

```python
url = ("https://financialmodelingprep.com/api/v4/economic?name=smoothedUSRecessionProbabilities&from=2019-07-10&to=2022-09-10&apikey=eef11245e039afd02560c03e5e86851d")
```

```python
data=get_jsonparsed_data(url)
print(json.dumps(data, indent=4))
```

```python
var = pd.DataFrame(data)
var
```
